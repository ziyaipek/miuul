```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blast_G_muris = pd.read_csv('/content/G_muris.blastn', sep='\t', header=None)
blast_S_salmonicida = pd.read_csv('/content/S_salmonicida.blastn', sep='\t',
header=None)
```

```python
blast_G_muris.columns = ['qseqid', 'sseqid', 'pident',
'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart',
'send', 'evalue', 'bitscore']
blast_G_muris.columns
blast_S_salmonicida.columns = ['qseqid', 'sseqid', 'pident',
'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart',
'send', 'evalue', 'bitscore']
```

**Histogram Plots**

```python
sns.histplot(data=blast_S_salmonicida , x='pident')
```

```python
sns.histplot(data=blast_G_muris, x='pident')
```

**Scatter Plots** \
*index vs bitscore*

```python
sns.scatterplot(data=blast_G_muris.reset_index(), x='index', y='bitscore', hue='pident')
```

```python
sns.scatterplot(data=blast_S_salmonicida.reset_index(), x='index', y='bitscore', hue='pident')
```

**Scatter Plots** \
*index vs length*

```python
sns.scatterplot(data=blast_G_muris.reset_index(), x='index', y='length', hue='pident')
```

```python
sns.scatterplot(data=blast_S_salmonicida.reset_index(), x='index', y='length', hue='pident')
```

**Heatmap**

```python
pivot_table_G_muris = pd.pivot_table(blast_G_muris, values = 'length', index = 'qseqid', columns = 'sseqid')
sns.heatmap(pivot_table_G_muris)
```

```python
pivot_table_S_salmonicida = pd.pivot_table(blast_S_salmonicida, values = 'length', index = 'qseqid', columns = 'sseqid')
sns.heatmap(pivot_table_S_salmonicida)
```
