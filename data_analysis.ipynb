```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blast_G_muris = pd.read_csv('/content/G_muris.blastn', sep='\t', header=None)
blast_G_muris
```

```python
blast_S_salmonicida = pd.read_csv('/content/S_salmonicida.blastn', sep='\t',
header=None)
blast_S_salmonicida
```

```python
blast_G_muris.columns = ['qseqid', 'sseqid', 'pident',
'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart',
'send', 'evalue', 'bitscore']
blast_G_muris.columns
blast_S_salmonicida.columns = ['qseqid', 'sseqid', 'pident',
'length', 'mismatch', 'gapopen', 'qstart', 'qend', 'sstart',
'send', 'evalue', 'bitscore']
```

```python
gi_gm = blast_G_muris[(blast_G_muris['sseqid'].str.contains('VSRT')) &
(blast_G_muris['qseqid'].str.contains('CM|VDLU'))]

gi_gm
```

```python
gi_ss = blast_S_salmonicida[(blast_S_salmonicida['sseqid'].str.contains('VSRT')) &
(blast_S_salmonicida['qseqid'].str.contains('CM|AUW'))]
gi_ss
```

```python
num_hits_gi_gm = len(gi_gm)
num_hits_gi_ss = len(gi_ss)
```

```python
print("Number of hits for Giardia intestinalis vs. Giardia muris: ", num_hits_gi_gm)
print("Number of hits for Giardia intestinalis vs. Spironucleus salmonicida: ", num_hits_gi_ss)
```

```python
# Create a figure with two subplots: one for the number of hits, one for the percent identity
fig, axs = plt.subplots(2, 1, figsize=(10, 10))
# Bar plot for the number of hits
axs[0].bar(['Giardia intestinalis vs. Giardia muris', 'Giardia intestinalis vs. Spironucleus salmonicida'], [num_hits_gi_gm, num_hits_gi_ss])
axs[0].set_ylabel('Number of hits')

# Scatter plot for the percent identity
axs[1].scatter(['Giardia intestinalis vs. Giardia muris']*len(gi_gm), gi_gm['pident'], alpha=0.5)
axs[1].scatter(['Giardia intestinalis vs. Spironucleus salmonicida']*len(gi_ss), gi_ss['pident'], alpha=0.5)
axs[1].set_ylabel('Percent Identity')
# Display the plots
plt.tight_layout()
plt.show()
```
